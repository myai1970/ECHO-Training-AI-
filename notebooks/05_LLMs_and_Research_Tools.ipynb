### 1.1 Basic Prompting Principles

### Core Principles:

1. **Be Specific**
   - âŒ Bad: "Tell me about aspirin"
   - âœ… Good: "Explain the mechanism of action of aspirin in reducing inflammation, including the specific enzymes it inhibits"

2. **Provide Context**
   - âŒ Bad: "Analyze this data"
   - âœ… Good: "As a chemistry researcher, analyze this spectroscopy data to identify the compound's functional groups"

3. **Specify Format**
   - âŒ Bad: "List synthesis methods"
   - âœ… Good: "List 5 synthesis methods for aspirin in a table with columns: Method, Yield %, Advantages, Disadvantages"

4. **Use Examples (Few-Shot Learning)**
   - Show the model what you want
   - Provide 1-3 examples of desired output

5. **Break Complex Tasks**
   - Divide large tasks into smaller steps
   - Use chain-of-thought prompting

## Bonus: Understanding SMILES Notation with AI

### What is SMILES?

**SMILES** = Simplified Molecular Input Line Entry System

It's a text-based notation for representing molecular structures that computers can easily read and process.

### SMILES Rules:

1. **Capital Letters = Atoms**
   - C = Carbon (aliphatic)
   - N = Nitrogen
   - O = Oxygen
   - S = Sulfur
   - P = Phosphorus

2. **Lowercase 'c' = Aromatic Carbon**
   - c = Carbon in aromatic ring (like benzene)
   - n, o, s = Aromatic nitrogen, oxygen, sulfur

3. **Bonds**
   - (nothing) = Single bond
   - = = Double bond
   - # = Triple bond

4. **Numbers = Ring Closures**
   - Same number appears twice to close a ring
   - c1ccccc1 = benzene (6-carbon aromatic ring)

5. **Parentheses ( ) = Branches**
   - Everything inside is a branch off the main chain

### Example SMILES:

```
CCO                    = Ethanol (CHâ‚ƒ-CHâ‚‚-OH)
CC(=O)O                = Acetic acid
c1ccccc1               = Benzene
CC(=O)Oc1ccccc1C(=O)O  = Aspirin
```

### Complex Example Breakdown:

Let's decode: `O=S(=O)(Nc1cccc(-c2cnc3ccccc3n2)c1)c1cccs1`

- **O=S(=O)** â†’ Sulfonyl group (SOâ‚‚)
- **(Nc1cccc...c1)** â†’ Amine attached to benzene ring
- **(-c2cnc3ccccc3n2)** â†’ Quinoxaline (fused aromatic rings)
- **(c1cccs1)** â†’ Thiophene (5-membered ring with sulfur)

### 2.1 Using Spacy for Basic NER

````python
import spacy
from spacy import displacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Scientific text example
text = """
Aspirin (acetylsalicylic acid) is a medication used to reduce pain, fever, or inflammation.
It was first isolated by Felix Hoffmann at Bayer in 1897. The compound inhibits
cyclooxygenase (COX) enzymes, particularly COX-1 and COX-2, which are involved in
prostaglandin synthesis. Clinical trials at Johns Hopkins University showed significant
reduction in cardiovascular events when patients took 81 mg daily.
"""

# Process text
doc = nlp(text)

# Extract entities
print("Named Entities Found:")
print("=" * 60)
for ent in doc.ents:
    print(f"{ent.text:30} | {ent.label_:15} | {spacy.explain(ent.label_)}")
````

````python
# Visualize entities
from IPython.core.display import HTML

html = displacy.render(doc, style="ent", jupyter=False)
display(HTML(html))
````

### 2.2 Using Transformers for Scientific NER (SciBERT)

````python
from transformers import pipeline

# Load NER pipeline (using a general model - in production, use domain-specific models)
ner = pipeline("ner", model="dslim/bert-base-NER", grouped_entities=True)

# Scientific abstract
abstract = """
A novel synthesis method for paracetamol was developed using green chemistry principles.
The reaction was catalyzed by palladium nanoparticles at 80Â°C for 2 hours.
Characterization was performed using NMR spectroscopy and mass spectrometry.
The yield was 95% with high purity confirmed by HPLC analysis.
"""

# Extract entities
entities = ner(abstract)

print("\nEntities Extracted:")
print("=" * 60)
for entity in entities:
    print(f"{entity['word']:30} | {entity['entity_group']:15} | Score: {entity['score']:.3f}")
````

### 2.3 Custom Entity Extraction with Regex

````python
import re

text = """
The reaction mixture contained 10 mg of compound A, 25 mL of ethanol,
and was heated to 80Â°C for 3 hours. The pH was adjusted to 7.4 using HCl.
Final yield: 85%. Melting point: 142-144Â°C.
Molecular formula: C9H8O4. MW: 180.16 g/mol.
"""

# Define patterns
patterns = {
    'Mass': r'\d+\.?\d*\s*(?:mg|g|kg)',
    'Volume': r'\d+\.?\d*\s*(?:mL|L|ÂµL)',
    'Temperature': r'\d+\.?\d*\s*Â°C',
    'Time': r'\d+\.?\d*\s*(?:hours?|mins?|seconds?)',
    'pH': r'pH\s*(?:of\s*)?\d+\.?\d*',
    'Percentage': r'\d+\.?\d*\s*%',
    'Molecular_Formula': r'C\d+H\d+(?:O\d+)?(?:N\d+)?',
}

print("Extracted Experimental Parameters:")
print("=" * 60)

for category, pattern in patterns.items():
    matches = re.findall(pattern, text)
    if matches:
        print(f"{category:20} : {', '.join(matches)}")
````

### 4.1 Literature Review Automation

````python
# Example: Extracting information from multiple abstracts
import pandas as pd

abstracts = [
    """
    Title: Green synthesis of silver nanoparticles using plant extracts
    Abstract: Silver nanoparticles were synthesized using Azadirachta indica leaf extract.
    The particles showed average size of 20-50 nm. Antimicrobial activity was tested
    against E. coli and S. aureus with MIC values of 10 Âµg/mL and 15 Âµg/mL respectively.
    """,
    """
    Title: Chemical synthesis of gold nanoparticles
    Abstract: Gold nanoparticles were synthesized by chemical reduction using HAuCl4
    and sodium citrate. Particle size was 15-25 nm. Catalytic activity for organic
    reactions showed 95% conversion rate at 80Â°C.
    """,
    """
    Title: Biological synthesis of copper nanoparticles
    Abstract: Copper nanoparticles synthesized using bacterial culture of Bacillus sp.
    Average size 30-40 nm. Showed photocatalytic degradation of methylene blue
    with 90% efficiency in 2 hours under UV light.
    """
]

# Extract structured data
data = []

for abstract in abstracts:
    # Extract particle type
    particle_match = re.search(r'(Silver|Gold|Copper) nanoparticles', abstract)
    particle_type = particle_match.group(1) if particle_match else 'Unknown'

    # Extract size
    size_match = re.search(r'(\d+)-(\d+)\s*nm', abstract)
    size_range = f"{size_match.group(1)}-{size_match.group(2)} nm" if size_match else 'Not reported'

    # Extract synthesis method type
    if 'plant' in abstract.lower() or 'leaf' in abstract.lower():
        method = 'Green/Plant-based'
    elif 'bacterial' in abstract.lower() or 'biological' in abstract.lower():
        method = 'Biological'
    elif 'chemical' in abstract.lower():
        method = 'Chemical'
    else:
        method = 'Other'

    data.append({
        'Particle Type': particle_type,
        'Size Range': size_range,
        'Synthesis Method': method
    })

# Create DataFrame
df_literature = pd.DataFrame(data)
print("Extracted Literature Data:")
print(df_literature.to_string(index=False))
````

### 4.2 Experimental Data Extraction

````python
# Extract experimental conditions from methods section
methods_text = """
Synthesis Procedure:
1. Heat 500 mL of water to 80Â°C
2. Add 100 mg of HAuCl4 and stir for 5 minutes
3. Add 50 mL of 1% sodium citrate solution
4. Maintain temperature at 80Â°C for 30 minutes
5. Cool to room temperature
6. pH was adjusted to 7.0 using NaOH
7. Centrifuge at 10000 rpm for 15 minutes
"""

# Extract parameters
extraction_patterns = {
    'Temperatures': r'\d+\s*Â°C',
    'Volumes': r'\d+\.?\d*\s*(?:mL|L)',
    'Masses': r'\d+\.?\d*\s*(?:mg|g)',
    'Times': r'\d+\.?\d*\s*(?:minutes?|hours?)',
    'pH': r'pH\s*(?:was\s*)?(?:adjusted\s*to\s*)?\d+\.?\d*',
    'Speed': r'\d+\s*rpm',
}

print("Extracted Experimental Conditions:")
print("=" * 60)

for param, pattern in extraction_patterns.items():
    matches = re.findall(pattern, methods_text)
    if matches:
        print(f"\n{param}:")
        for match in set(matches):  # Use set to avoid duplicates
            print(f"  - {match}")
````

````python
# Approach 1: HuggingFace Inference API (Simple, No GPU Needed)
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

def generate_image_with_api(prompt, api_token):
    """
    Generate image using HuggingFace Inference API

    Args:
        prompt: Text description of image to generate
        api_token: Your HuggingFace API token

    Returns:
        PIL Image object
    """
    API_URL = "https://api-inference.huggingface.co/models/stabilityai/stable-diffusion-2-1"
    headers = {"Authorization": f"Bearer {api_token}"}

    payload = {
        "inputs": prompt,
        "parameters": {
            "num_inference_steps": 30,  # Lower for faster generation
            "guidance_scale": 7.5
        }
    }

    response = requests.post(API_URL, headers=headers, json=payload)

    if response.status_code == 200:
        image = Image.open(BytesIO(response.content))
        return image
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None

# IMPORTANT: Get your token from https://huggingface.co/settings/tokens
# Replace 'YOUR_TOKEN_HERE' with your actual token
HF_API_TOKEN = "YOUR_TOKEN_HERE"  # <-- REPLACE THIS

# Scientific image prompts
scientific_prompts = [
    "molecular structure of aspirin, detailed chemistry diagram, white background",
    "laboratory beaker with blue chemical solution, scientific photography",
    "chemical synthesis glassware setup, round bottom flask, condenser, scientific illustration"
]

print("Image Generation with HuggingFace API")
print("=" * 80)
print("NOTE: This requires a HuggingFace API token")
print("Get yours at: https://huggingface.co/settings/tokens")
print("=" * 80)

# Example (will only work if you provide your token)
if HF_API_TOKEN != "YOUR_TOKEN_HERE":
    prompt = scientific_prompts[0]
    print(f"\nGenerating: {prompt}")

    image = generate_image_with_api(prompt, HF_API_TOKEN)

    if image:
        plt.figure(figsize=(8, 8))
        plt.imshow(image)
        plt.axis('off')
        plt.title(prompt[:50] + "...")
        plt.tight_layout()
        plt.show()

        print("[OK] Image generated successfully!")
    else:
        print("[X] Failed to generate image. Check your API token.")
else:
    print("\n[!] Please add your HuggingFace API token to run this example")
    print("[!] For now, showing example prompts you can use:")
    print("\n".join(f"  {i+1}. {p}" for i, p in enumerate(scientific_prompts)))
````

````python
# Template-based approach - Simple, reliable, no AI needed
class ResearchReportTemplate:
    """Generate research reports using templates"""

    def __init__(self, title, author, date):
        self.title = title
        self.author = author
        self.date = date

    def generate_abstract(self, objective, methods, results, conclusion):
        """Generate abstract section"""
        return f"""**Abstract**

**Objective:** {objective}

**Methods:** {methods}

**Results:** {results}

**Conclusion:** {conclusion}
"""

    def generate_methods(self, materials, procedure, analysis):
        """Generate methods section"""
        return f"""**Methods**

**Materials:**
{materials}

**Procedure:**
{procedure}

**Analysis:**
{analysis}
"""

    def generate_results(self, findings, data_summary):
        """Generate results section"""
        return f"""**Results**

**Key Findings:**
{findings}

**Data Summary:**
{data_summary}
"""

    def generate_full_report(self, abstract_data, methods_data, results_data):
        """Generate complete report"""
        report = f"""# {self.title}

**Author:** {self.author}
**Date:** {self.date}

---

"""
        report += self.generate_abstract(**abstract_data)
        report += "\n---\n\n"
        report += self.generate_methods(**methods_data)
        report += "\n---\n\n"
        report += self.generate_results(**results_data)

        return report

# Example usage
report_gen = ResearchReportTemplate(
    title="Synthesis and Characterization of Silver Nanoparticles",
    author="DOST-ITDI Research Team",
    date="November 2025"
)

# Define report content
abstract_data = {
    'objective': "Synthesize silver nanoparticles using green chemistry methods",
    'methods': "Plant extract reduction method using Azadirachta indica",
    'results': "Spherical nanoparticles (25 nm average size) with strong antimicrobial activity",
    'conclusion': "Successful green synthesis of bioactive silver nanoparticles"
}

methods_data = {
    'materials': """- Azadirachta indica (neem) leaves
- Silver nitrate (AgNO3)
- Distilled water
- Standard laboratory glassware""",
    'procedure': """1. Collect and wash fresh neem leaves
2. Prepare leaf extract by boiling in distilled water
3. Filter extract and cool to room temperature
4. Add silver nitrate solution dropwise
5. Incubate at room temperature for 2 hours
6. Centrifuge and wash nanoparticles
7. Dry at 60 degrees C""",
    'analysis': """- UV-Vis spectroscopy for optical properties
- TEM for size and morphology
- XRD for crystal structure
- Antimicrobial assay against E. coli and S. aureus"""
}

results_data = {
    'findings': """- UV-Vis peak at 420 nm confirmed nanoparticle formation
- TEM showed spherical particles, 20-30 nm diameter
- XRD confirmed face-centered cubic silver structure
- MIC values: E. coli (10 Âµg/mL), S. aureus (15 Âµg/mL)""",
    'data_summary': """Average particle size: 25 Â± 5 nm
Zeta potential: -28 mV (stable)
Yield: 85%
Antimicrobial efficacy: High against both bacteria"""
}

# Generate report
full_report = report_gen.generate_full_report(abstract_data, methods_data, results_data)

print("Generated Research Report:")
print("="*80)
print(full_report)
print("="*80)
````
